```python
def coefs_1d(N,N0,lab) :
    return vector([ var(lab+'%s'%i) for i in range(N0,N0+N) ])

def poly_1d(N,coefs) :
    var('x')
    return sum( vector([ coefs[i]*x^i for i in range(N) ]) )

def compute_coefs(N,p) :
    return vector([ diff(p,x,i).subs({x:0})/factorial(i) for i in range(N) ])

def integrate_poly(N,p,x1,x2) :
    var('a')
    a = compute_coefs(N,p)
    return sum( vector([ a[i]*(x2**(i+1) - x1**(i+1)) / (i+1) for i in range(N) ]) )

def c_scalar_pow2(retlab,val,coeflab) :
    import re
    code = ""
    s = retlab+"="
    #Remove spaces
    s = s+str(val).replace(' ','')
    #Replace x^2 with (x*x)
    s = re.sub("([a-zA-Z()0-9_]*)\^2","(\\1*\\1)",s,0,re.DOTALL)
    #Replace coeficients (e.g., c1) with parentheses (e.g., c(1))
    s = re.sub(coeflab+"([0-9]+)",coeflab+"[\\1]",s, 0, re.DOTALL)
    #Break up code to the given wraplen for line length
    #s = re.sub("&\n&$","",re.sub("(.{"+str(wraplen)+"})", "\\1&\n&", s, 0, re.DOTALL))
    #Add new line to the end.
    code = code + s + ";\n"
    return code

def force_fp(expr,prec) :
    R = RealField( prec )
    expr = expr * R(2.) / R(2.)
    return expr

for N in range(2,10) :
    p = poly_1d( N , coefs_1d(N,0,'a') )
    expr = integrate_poly( N , sum( vector([ p.diff(x,i)^2 for i in range(1,N) ]) ) , -1/2 , 1/2 )
    print( c_scalar_pow2( 'tv' , force_fp(expr,129) , 'a' ))
```
